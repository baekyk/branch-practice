```python
  import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = '/content/drive/MyDrive/Colab Notebooks/titanic/'

train = pd.read_csv(base_path + 'train.csv')
test = pd.read_csv(base_path + 'test.csv')
submission = pd.read_csv(base_path + 'gender_submission.csv')
```

```python
train.info()
```

# Data Preprocessing

```python
train[train.isnull().any(axis=1)]
```

```python
# train[train['Embarked'].isnull()]

# train.Embarked.value_counts() --> 'S'

# train[(train.Pclass == 1) & (train.Sex == 'female')].Embarked.value_counts() -->'s'

train.loc[train.Embarked.isnull(), 'Embarked'] = "S"
train.info()
```

```python
train = train.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
train.info()
```

```python
train = train.fillna(train.Age.mean())
```

```python
train.info()
```

```python
train_OHE = pd.get_dummies(train, columns=["Sex", "Embarked"])
```

```python
train_OHE
```

```python
from sklearn.preprocessing import MinMaxScaler

X = train_OHE.drop(columns='Survived')
y = train_OHE.Survived

scaler = MinMaxScaler()
# scaler.fit()
# scaler.transform()
temp = scaler.fit_transform(X.loc[:,["Age", "Fare"]])

X["Age"] = temp[:,0]
X["Fare"] = temp[:,1]
X
```

# Training

```python
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
```

```python
clf = SGDClassifier()
clf2 = LogisticRegression()
clf3 = DecisionTreeClassifier()
clf4 = RandomForestClassifier()

clf.fit(X, y)
clf2.fit(X, y)
clf3.fit(X, y)
clf4.fit(X, y)

pred = clf.predict(X)
pred2 = clf2.predict(X)
pred3 = clf3.predict(X)
pred4 = clf4.predict(X)
```

```python
print("1. Linear classifier, Accuracy for training : %.4f" % accuracy_score(y, pred))
print("2. Logistic Regression, Accuracy for training : %.4f" % accuracy_score(y, pred2))
print("3. DecisionTree, Accuracy for training : %.4f" % accuracy_score(y, pred3))
print("4. RandomForest, Accuracy for training : %.4f" % accuracy_score(y, pred4))
```
